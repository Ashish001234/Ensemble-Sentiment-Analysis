```python
from transformers import AutoModelForSequenceClassification
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoConfig
import numpy as np
from scipy.special import softmax
import csv
import urllib.request

def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment-latest"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
text = preprocess(text)
encoded_input = tokenizer(text, return_tensors='pt')
output = model(**encoded_input)
scores = output[0][0].detach().numpy()
scores = softmax(scores)

lst1 = []
ranking = np.argsort(scores)
ranking = ranking[::-1]
for i in range(scores.shape[0]):
    l = config.id2label[ranking[i]]
    s = scores[ranking[i]]
    lst1.append(s)
    print(f"{i+1}) {l} {np.round(float(s), 4)}")

print(lst1)
```

```python
task='sentiment'
MODEL = f"mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis"

tokenizer = AutoTokenizer.from_pretrained(MODEL)

labels=[]
mapping_link = f"https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/{task}/mapping.txt"
with urllib.request.urlopen(mapping_link) as f:
    html = f.read().decode('utf-8').split("\n")
    csvreader = csv.reader(html, delimiter='\t')
labels = [row[1] for row in csvreader if len(row) > 1]

# PT
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
model.save_pretrained(MODEL)


text = preprocess(text)
encoded_input = tokenizer(text, return_tensors='pt')
output = model(**encoded_input)
scores = output[0][0].detach().numpy()
scores = softmax(scores)

lst2 =[]
ranking = np.argsort(scores)
ranking = ranking[::-1]
for i in range(scores.shape[0]):
    l = labels[ranking[i]]
    s = scores[ranking[i]]
    lst2.append(s)
    print(f"{i+1}) {l} {np.round(float(s), 4)}")

temp = lst2[0]
lst2[0] = lst2[1]
lst2[1] = temp
print(lst2)
```

```python
MODEL = f"avichr/heBERT_sentiment_analysis"

tokenizer = AutoTokenizer.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL)

# PT
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
model.save_pretrained(MODEL)

text = preprocess(text)
encoded_input = tokenizer(text, return_tensors='pt')
output = model(**encoded_input)
scores = output[0][0].detach().numpy()
scores = softmax(scores)

lst5 = []
ranking = np.argsort(scores)
ranking = ranking[::-1]
for i in range(scores.shape[0]):
    l = config.id2label[ranking[i]]
    s = scores[ranking[i]]
    lst5.append(s)
    print(f"{i+1}) {l} {np.round(float(s), 4)}")
print(lst5)
```

```python
from collections import Counter

sentiment_scores_lists = [
    lst1, lst2, lst3, lst4, lst5
]

labels = ["positive", "neutral", "negative"]

consensus_sentiment = []

for scores in sentiment_scores_lists:
    max_index = np.argmax(scores)
    max_label = labels[max_index]
    consensus_sentiment.append(max_label)

sentiment_counts = Counter(consensus_sentiment)

final_sentiment = None
max_accuracy = 0

for sentiment, count in sentiment_counts.items():
    if count >= 2:
        indices = [i for i, s in enumerate(consensus_sentiment) if s == sentiment]
        for index in indices:
            accuracy = sentiment_scores_lists[index][labels.index(sentiment)]
            if accuracy > max_accuracy:
                max_accuracy = accuracy
                final_sentiment = sentiment

if final_sentiment is not None:
    print(f"The final consensus sentiment of '{text}' is: {final_sentiment}")
else:
    print("No consensus reached among the models.")
```
